###Function to perform hydraulic regression

```python
def harmonic_analysis(station,pharmonics,t,v):
    npfreq = len(pharmonics)
    npsamples = v.shape[0]
    nf2 = 2 * npfreq
    nf21, nf22 = nf2+1, nf2+2
    aa = zeros((nf22, nf22), float)
    ci = zeros(npfreq, float)
    si = zeros(npfreq, float)
    x = zeros(nf21, float)
    trig = zeros(nf22, float)
    amplitude = zeros(npfreq, float)
    theta = zeros(npfreq, float)
    trig[0] = 1.
    for i in xrange(npsamples):
        for j in xrange(1,nf2,2):
            jh = (j-1) / 2
            period = pharmonics[jh][1] * 60. * 60.
            if period==0.0: print period
            sigma = 2. * pi / period
            arc = sigma * t[i]
            j1 = j + 1
            trig[j] = cos(arc)
            trig[j1] = sin(arc)
        trig[nf22-1] = v[i]
        # fill matrix
        for i1 in xrange(nf21):
            for j1 in xrange(i1,nf22):
                aa[i1,j1] = aa[i1,j1] + trig[i1] * trig[j1]
                aa[j1,i1] = aa[i1,j1]
    # scale diagonal and off diagonals
    for i in xrange(nf21):
        diagonal = aa[i,i]
        aa[i,i] = 1.0
        for j in xrange(nf22):
            if (j != i):
                aa[i,j] = aa[i,j] / diagonal
    # solve equations
    determinant = splin.det(aa[:nf21,:nf21])
    if abs(determinant) < 1.e-20: 
        print 'singular or ill-conditioned matrix'
        print 'determinant = (0}'.format(determinant)
    ainv = splin.inv(aa[:nf21,:nf21])
    # calculate x
    for i in xrange(nf21):
        x[i] = 0.
        for j in xrange(nf21):
            x[i] += ainv[i,j] * aa[j,nf22-1]
    # calculate amplitude and phase
    for i in xrange(npfreq):
        i2 = (i * 2) + 1
        ci[i] = x[i2]
        si[i] = x[i2+1]
        amplitude[i] = sqrt(ci[i] * ci[i] + si[i] * si[i])
        theta[i] = 360. - arctan2(si[i], ci[i]) * 180. / pi #degrees
    #--write the data
    print station
    for idx,[c,period,pa,ptheta] in enumerate(pharmonics):
        print '  {0:5s} {1:15.7g} {2:15.7g} deg.'.format(c, amplitude[idx], theta[idx])
    return amplitude, theta   
```

###Read and process data

```python
#function to parse data and normalize it
def parse_data(fname):
    data_dtype = dtype([('datetime','a20'),('head','f4')])
    raw = loadtxt(fname, delimiter=',', dtype=data_dtype, comments='#')
    dout = []
    for [d,v] in raw:
        dt = datetime.datetime.strptime(d, '%m/%d/%Y %H:%M:%S')
        dout.append([dt,v])
    dout = array(dout)
    dout[:,1] -= average(dout[:,1])
    return dout
```

```python
data_dir = os.path.join('data')
tide_file = 'DE_Tide.csv'
cat_file = 'DE_Catalog.csv'
tide = ['tide',0.0,parse_data(os.path.join(data_dir,tide_file))]
time_min = tide[2][0,0]
time_max = tide[2][-1,0]
dt_sec = (tide[2][1,0] - tide[2][0,0]).total_seconds()
fo = open(os.path.join(data_dir,cat_file), 'r')
gw_data = []
for line in fo:
    if len(line) < 1:
        break
    tc = line.strip().replace(' ','').split(',')
    station = tc[0]
    x = float(tc[2])
    t = parse_data(os.path.join(data_dir,tc[1]))
    gw_data.append([station,x,t])
    time_min = min(time_min, t[0,0])
    time_max = max(time_max, t[-1,0])
fo.close()
#--create time_all
tstep = datetime.timedelta(seconds=dt_sec)
dt = time_min
time_all = []
while dt < time_max:
    time_all.append(dt)
    dt += tstep
time_all = array(time_all)
#--plot raw normalized data
ax = subplot(1,1,1)
t = tide[2]
ax.plot(t[:,0], t[:,1], linewidth=1.0)
for [station,x,gd] in gw_data:
    ax.plot(gd[:,0], gd[:,1], linewidth=0.5)
set_daysFmt(ax)
show()
```

###Function to intersect two data sets and calculate for a time series data set

```python
def pair_data(v,base):
    a = pd.DataFrame(v[:,1],index=v[:,0])
    b = pd.DataFrame(base[:,1],index=base[:,0])
    c = pd.concat([a,b], axis=1, join='inner')
    t = c.index.to_pydatetime()
    pairv = array(zip(t, c.values[:,0]))
    pbase = array(zip(t, c.values[:,1]))
    return pairv, pbase
def calculate_t(dt):
    total_sec = (dt[-1] - dt[0]).total_seconds()
    nsamples = dt.shape[0]
    t = linspace(0,total_sec,nsamples)
    return t
```

###Function to recreate time series, calculate l2 norm, and calculate error

```python
def create_signal(pharmonics,t):
    nsamples = t.shape
    v = zeros(nsamples, float)
    for idx,[c,period,a,theta,te,dtheta] in enumerate(pharmonics):
        period *= 60. * 60.
        fr = 1. / period
        omega = 2. * pi * fr
        theta *= pi / 180.
        phase = dtheta * pi / 180.
        v += te * a * cos(omega*t + theta + phase)
    return v
def signal_error(pharmonics,obsv):
    terr = calculate_t(obsv[:,0])
    simv = create_signal(pharmonics,terr)
    resid = subtract(simv,obsv[:,1])
    me = average(resid)
    rmse = std(resid)
    return me, rmse
```

###Process tidal data

```python
#--process tide
station = 'tide'
tide_data = tide[2]
t = calculate_t(tide_data[:,0])
tide_alpha, tide_theta = harmonic_analysis(station,harmonics,t,tide_data[:,1])
#--plot simulated and observed tide
t = calculate_t(time_all)
harmonic_results = []
for idx,[c,period,a,tv] in enumerate(harmonics):
    harmonic_results.append([c, period, tide_alpha[idx], tide_theta[idx], 1.0, 0.0])
dsec = (time_min-tide_data[0,0]).total_seconds()
t += dsec
vr = create_signal(harmonic_results,t)
me, rmse = signal_error(harmonic_results,tide_data)
ctxt = '{0} - ME: {1:7.2g} meters, RMSE: {2:7.2g} meters'.format('Tide', me, rmse)
ax = subplot(1,1,1)
ax.plot(time_all, vr, linewidth=1.0)
ax.plot(tide_data[:,0::5], tide_data[:,1::5], linewidth=0.0, marker='x', markersize=3)
ax.text(0.0, 1.01, ctxt, ha='left', va='bottom', size=8, transform=ax.transAxes)
set_daysFmt(ax)
show()
```

###Determine TE and dtheta

```python
station_te = []
station_dtheta = []
for [station,x,gd] in gw_data:
    pairv, pb = pair_data(gd,tide_data)
    t = calculate_t(pairv[:,0])
    alpha, theta = harmonic_analysis(station,harmonics,t,pairv[:,1])
    a2, t2 = harmonic_analysis(station,harmonics,t,pb[:,1])
    te = []
    dtheta = []
    # write the station data
    print station
    for idx,[c,period,ta,tv] in enumerate(harmonics):
        v1 = alpha[idx]/a2[idx]
        v2 = theta[idx]-t2[idx]
        te.append(v1)
        dtheta.append(v2)
        #--write the data
        print '  {0:5s} {1:15.7g} {2:15.7g} deg.'.format(c, v1, v2)
    station_te.append(te)
    station_dtheta.append(dtheta)        
```

###Plot reconstructed data

```python
npcol = 2
nprow = 1 + (1 + len(gw_data)) / npcol
# calculate simulated tidal value
t = calculate_t(time_all)
harmonic_results = []
for idx,[c,period,a,tv] in enumerate(harmonics):
    harmonic_results.append([c, period, tide_alpha[idx], tide_theta[idx], 1.0, 0.0])
dsec = (time_min-tide[2][0,0]).total_seconds()
t += dsec
vr = create_signal(harmonic_results,t)
me, rmse = signal_error(harmonic_results,tide[2])
ctxt = '{0} - ME: {1:7.2g} meters, RMSE: {2:7.2g} meters'.format('Tide', me, rmse)
ifig = 1
figure(num=None, figsize=(10, 8), dpi=300, facecolor='w', edgecolor='k')
ax = subplot(nprow,npcol,ifig)
ax.plot(time_all, vr, linewidth=1.0)
ax.plot(tide[2][:,0], tide[2][:,1], linewidth=0.0, marker='x', markersize=2)
ax.text(0.0, 1.01, ctxt, ha='left', va='bottom', size=8, transform=ax.transAxes)
set_daysFmt(ax)
#--process each groundwater file
for istat,[station,x,gd] in enumerate(gw_data):
    ifig += 1
    t -= dsec
    harmonic_results = []
    for idx,[c,period,a,tv] in enumerate(harmonics):
        harmonic_results.append([c, period, tide_alpha[idx], tide_theta[idx], station_te[istat][idx], station_dtheta[istat][idx]])
    dsec = (time_min-gd[0,0]).total_seconds()
    t += dsec
    vr = create_signal(harmonic_results,t)
    me, rmse = signal_error(harmonic_results,gd)
    ctxt = '{0} - ME: {1:7.2g} meters RMSE: {2:7.2g} meters'.format(station, me, rmse)
    ax = subplot(nprow,npcol,ifig)
    ax.plot(time_all, vr, linewidth=1.0)
    ax.plot(gd[:,0], gd[:,1], linewidth=0.0, marker='x', markersize=2)
    ax.text(0.0, 1.01, ctxt, ha='left', va='bottom', size=8, transform=ax.transAxes)
    set_daysFmt(ax)
show()
```
