###Function to perform hydraulic regression

```python
def harmonic_analysis(station,pharmonics,time_sec,stage,print_results=False):
    npfreq = len(pharmonics)
    npsamples = stage.shape[0]
    nf2 = 2 * npfreq
    nf21, nf22 = nf2+1, nf2+2
    aa = zeros((nf22, nf22), float)
    ci = zeros(npfreq, float)
    si = zeros(npfreq, float)
    x = zeros(nf21, float)
    trig = zeros(nf22, float)
    amplitude = zeros(npfreq, float)
    theta = zeros(npfreq, float)
    trig[0] = 1.
    for i in xrange(npsamples):
        for j in xrange(1,nf2,2):
            jh = (j-1) / 2
            period = pharmonics[jh][1] * 60. * 60.
            if period==0.0: print period
            fr = 1. / period
            omega = 2. * pi * fr
            arc = omega * t[i]
            j1 = j + 1
            trig[j] = cos(arc)
            trig[j1] = sin(arc)
        trig[nf22-1] = stage[i]
        # fill matrix
        for i1 in xrange(nf21):
            for j1 in xrange(i1,nf22):
                aa[i1,j1] = aa[i1,j1] + trig[i1] * trig[j1]
                aa[j1,i1] = aa[i1,j1]
    # scale diagonal and off diagonals
    for i in xrange(nf21):
        diagonal = aa[i,i]
        aa[i,i] = 1.0
        for j in xrange(nf22):
            if (j != i):
                aa[i,j] = aa[i,j] / diagonal
    # solve equations
    determinant = splin.det(aa[:nf21,:nf21])
    if abs(determinant) < 1.e-20: 
        print 'singular or ill-conditioned matrix'
        print 'determinant = {0}'.format(determinant)
    ainv = splin.inv(aa[:nf21,:nf21])
    # calculate x
    for i in xrange(nf21):
        x[i] = 0.
        for j in xrange(nf21):
            x[i] += ainv[i,j] * aa[j,nf22-1]
    # calculate amplitude and phase
    harmonic_results = []
    for i in xrange(npfreq):
        i2 = (i * 2) + 1
        ci[i] = x[i2]
        si[i] = x[i2+1]
        amplitude[i] = sqrt(ci[i] * ci[i] + si[i] * si[i])
        ttheta = arctan2(si[i], ci[i]) #* 180. / pi #degrees
        if ttheta < 0.:
            ttheta += 2. * pi
        theta[i] = rad2deg(ttheta)
        harmonic_results.append([pharmonics[i][0], pharmonics[i][1], amplitude[i], theta[i]])
    #--write the data
    if print_results:
        print_hr(station,harmonic_results)
    return amplitude, theta   

def print_hr(station,pharmonics,difference=False):
    print station
    if difference:
        print '  {0:5s} {1:^16s} {2:^16s}'.format('', 'Tidal Effic.', 'Phase Diff.')
    else:
        print '  {0:5s} {1:^16s} {2:^16s}'.format('', 'Amplitude', 'Phase')
    print '  {0:5s} {1:^16s} {2:^16s}'.format('Name', 'meters', 'degrees')
    print '  {0:5s} {1:^16s} {2:^16s}'.format('-----', '----------------', '----------------')
    for idx,[c,period,pa,ptheta] in enumerate(pharmonics):
        print '  {0:5s} {1: 16.3f} {2: 16.1f}'.format(c, pa, ptheta)
    print ''
```

###Read and process data

```python
#function to parse data and normalize it
def parse_data(fname):
    data_dtype = dtype([('datetime','a20'),('head','f4')])
    raw = loadtxt(fname, delimiter=',', dtype=data_dtype, comments='#')
    dtime = []
    stage = []
    for [d,v] in raw:
        dt = datetime.datetime.strptime(d, '%m/%d/%Y %H:%M:%S')
        dtime.append(dt)
        stage.append(v)
    dtime = array(dtime)
    stage = array(stage)
    stage -= average(stage)
    return dtime, stage
```

###Function to intersect two data sets and calculate for a time series data set

```python
def pair_data(pdatetime_in,pv_in,pdatetime_base,pv_base):
    a = pd.DataFrame(pv_in,index=pdatetime_in)
    b = pd.DataFrame(pv_base,index=pdatetime_base)
    c = pd.concat([a,b], axis=1, join='inner')
    pdatetime_out = c.index.to_pydatetime()
    return pdatetime_out, c.values[:,0], c.values[:,1]
def calculate_t(datetime_in):
    total_sec = (datetime_in[-1] - datetime_in[0]).total_seconds()
    nsamples = datetime_in.shape[0]
    t_out = linspace(0.,float(total_sec),nsamples)
    return t_out
```

###Function to recreate time series, calculate l2 norm, and calculate error

```python
def create_signal(pharmonics,t_cs):
    nsamples = t_cs.shape
    v_out = zeros(nsamples, float)
    for idx,[c,period,a,theta,te,phase] in enumerate(pharmonics):
        period *= 60. * 60.
        fr = 1. / period
        omega = 2. * pi * fr
        theta *= pi / 180.
        phase *= pi / 180.
        v_out += te * a * cos(omega*t_cs - theta - phase)
    return v_out
def signal_error(pharmonics,datetime_se,obsv):
    t_se = calculate_t(datetime_se)
    simv = create_signal(pharmonics,t_se)
    resid = subtract(simv,obsv)
    me = average(resid)
    rmse = std(resid)
    return me, rmse
```

###Determine TE and dtheta

```python
#--minimize error
class fit_te:
    def __init__(self, station, class_harmonics, t, obs, method='nelder-mead'):
        self.station = station
        self.class_harmonics = class_harmonics
        self.nfreq = len(self.class_harmonics)
        self.t = t
        self.obs = obs
        self.method = method
        self.te = empty(self.nfreq, float)
        self.te.fill(0.9)
        self.dtheta = empty(self.nfreq, float)
        #self.dtheta.fill(0.1)
        self.calculate_dtheta()
    
    def fill_pharmonics(self, x):
        #self.te = x[:nfreq]
        #self.dtheta = x[nfreq:]
        self.te = x
        self.calculate_dtheta()
        pharmonics = []
        for idx,[c,period,a,tv] in enumerate(self.class_harmonics):
            pharmonics.append([c, period, a, tv, self.te[idx], self.dtheta[idx]])
        return pharmonics

    def calculate_dtheta(self):
        for idx,[c,period,a,tv] in enumerate(self.class_harmonics):
            period *= 60. * 60.
            fr = 1. / period
            omega = 2. * pi * fr
            ttheta = -1. * log(self.te[idx]) / omega
            self.dtheta[idx] = ttheta
        
    def calculate_l2norm(self, x):
        pharmonics = self.fill_pharmonics(x)
        sim_out = create_signal(pharmonics,self.t)
        resid = subtract(self.obs,sim_out)
        return splin.norm(resid)

    def te_analysis(self):
        #x0 = append(self.te, self.dtheta)
        x0 = self.te[:]
        res = sciopt.minimize(self.calculate_l2norm, x0, method=self.method,\
                             tol=1e-6, options={'disp': True, 'maxiter': 1000})
        #self.te = res.x[:nfreq]
        #self.dtheta = res.x[nfreq:]
        self.te = res.x
        self.calculate_dtheta()
        return self.te, self.dtheta
```

```python
station_te = []
station_dtheta = []
for [station, x, gd, gv] in gw_data:
    pdatetime, pv, pb = pair_data(gd, gv, tide_datetime, tide_v)
    t = calculate_t(pdatetime)
    #alpha, theta = harmonic_analysis(station,harmonics,t,pb)
    alpha, theta = harmonic_analysis(station,harmonics,t,pv)
    a2, t2 = harmonic_analysis(station,harmonics,t,pb)
    harmonic_results = []
    for idx,[c,period,ta,tv] in enumerate(harmonics):
        harmonic_results.append([c, period, alpha[idx], theta[idx]])
    #fit_tidal = fit_te(station, harmonic_results, t, pv)
    #te, dtheta = fit_tidal.te_analysis()
    te = []
    dtheta = []
    # write the station data
    harmonic_results = []
    for idx,[c,period,ta,tv] in enumerate(harmonics):
        v1 = alpha[idx]/a2[idx]
        v2 = theta[idx]-t2[idx]
        te.append(v1)
        dtheta.append(v2)
        harmonic_results.append([c, period, v1, v2])
        #harmonic_results.append([c, period, te[idx], dtheta[idx]])
    station_te.append(te)
    station_dtheta.append(dtheta)
    print_hr(station,harmonic_results,difference=True)
```
